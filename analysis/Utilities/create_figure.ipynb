```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
arr=np.array([[84,37.3],[16,62.2],[16,54.5],[33,72.9],[16,83.5],[16,72.5]])
df=pd.DataFrame(arr,columns=["img/cow","Acc"],index=["Takamatu","Nakamura","Hutiwaki","F.Okura","Goto_Resnet","Goto_GBDT"])
```

```python
df.index[0]
```

```python
import matplotlib.cm as cm 
fig = plt.figure(figsize=(7,5))

ax = fig.add_subplot(111)
i=0
colors = cm.rainbow(np.linspace(0, 1, df.shape[0]))
for dataset,color in zip(df["Acc"],colors):
    plt.scatter(df["img/cow"][i],dataset,color=color,label=df.index[i])
    
    i+=1
plt.title("Cow Identification(3D)")
plt.ylabel("Accuracy",fontsize=13)
plt.xlabel("Image/Cow",fontsize=13)
ax.set_yticks([0, 20, 40,60,80,100])
plt.legend(loc='upper right',fontsize=13)
plt.show()
fig.savefig("img.png")
```
